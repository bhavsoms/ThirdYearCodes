```python
#importing necessary libraries
from numpy import array, random, dot 
from random import choice
from pylab import ylim,plot
from matplotlib import pyplot as pyplot
import numpy as np
```

```python
step_function=lambda x:0 if x<0 else 1
```

OR GATE- PERCEPTRON RULE

```python
#input values of OR with bias=1
training_dataset=[
    (array([0,0,1]),0),
    (array([0,1,1]),1),
    (array([1,0,1]),1),
    (array([1,1,1]),1)
]
```

```python
weights=random.rand(3)
error=[]
learning_rate=0.2
n=100
```

```python
for j in range(n):
    x, expected = choice(training_dataset)#Randomly selects a sample (x) from the training dataset along with its corresponding expected output (expected).
    result = dot(weights, x)#calculates dot product between weights and input
    err = expected - step_function(result)
    error.append(err)
    weights += learning_rate *err *x
for x, _ in training_dataset:
    result = dot(x, weights)
    print('{}: {} -> {}'.format(x[:2], result, step_function(result)))
    #: Prints the input and the corresponding output of the perceptron.
ylim([-1,1])
plot(error)
```

AND GATE-PERCEPTRON LEARNING

```python
training_dataset1=[
    (array([0,0,1]),0),
    (array([0,1,1]),0),
    (array([1,0,1]),0),
    (array([1,1,1]),1)
]
```

```python
weight=random.rand(3)
error1=[]
learning_rate1=0.2
n=100
```

```python
for j in range(n):
    x, expected = choice(training_dataset1)
    result1 = dot(weight, x)
    err1 = expected - step_function(result1)
    error1.append(err1)
    weight += learning_rate1 * err1 * x

for x, _ in training_dataset1:
    result1 = dot(weight, x)
    print('{}: {} -> {}'.format(x[:2], result1, step_function(result1)))

ylim(-1, 1)
plot(error1)
    
```
