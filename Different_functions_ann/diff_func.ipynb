```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

Binary Activation Function

Heaviside in Numpy for binary step function, x2=1

```python
#Binary Step Function
def binaryStep(x):
    "'It returns '0' if the input is less than zero otherwise it return 1"
    return np.heaviside(x,1)#x2=1
x = np.linspace(-10, 10)  
plt.plot(x, binaryStep(x))  
plt.axis('tight')  
plt.title('Activation Function :BinaryStep')  
plt.show()  
```

Linear Activation Function

```python
def linear(x):
    "y=f(x) it returns the input as it is"
    return x
x=np.linspace(-10,10)
plt.plot(x,linear(x))
plt.axis('tight')
plt.title('Activation Linear Function')
plt.show()
```

Sigmoid Activation Function

```python
def sigmoid(x):
    "It returns 1/(1+exp(-x)) when the value lies between 0 and 1"
    return 1/(1+np.exp(-x))
x=np.linspace(10,-10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function: Sigmoid')
plt.show()
```

Bipolar Sigmoidal Function

```python
def bipolar_sigmoidal(x):
    "It returns [-1+(2/(1+exp(-x)))] where the value lies between -1 and 1"
    return -1+2/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,bipolar_sigmoidal(x))
plt.axis('tight')
plt.title('Activation Function: Bipolar Sigmoidal')
plt.show()
```

Tanh Activation Function

```python
def tanh(x):
    "It returns the value (1-exp(-x)/1+exp(-x)) for x between -1 and 1"
    return np.tanh(x)
x= np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Activation Function:Tanh')
plt.show()
```

ReLU Activation Function

```python
def ReLU(x):
    "It returns 0 if the input is less than 0 else it return the given input"
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x=np.linspace(-10,10)
plt.plot(x, ReLU(x))
plt.axis('tight')
plt.title('Activation Function: RELU')
plt.show()
```

Leaky RELU

```python
'It returns value smaller to zero if the input is less than 0 by multipluing it by alpha else returns input '
alpha= 0.01
def leakyrelu (x):
    return np.maximum(alpha*x,x)
x=np.linspace(-10,10)
plt.plot(x,leakyrelu(x))
plt.axis('tight')
plt.title('Activation Function: Leaky Relu')
plt.show()
```

Soft Max Activation Function

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
x=np.linspace(-10,10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title('SoftMax Activation Function')
plt.show()
```
